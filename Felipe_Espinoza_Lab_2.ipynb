```python
def outer():
  x = 10
  def inner():
    return x
    return inner

  f = outer()
  print(f())
```

```python
def deco(fn):
  def wrapper():
    print("before")
    fn()
    print("after")
  return wrapper # The deco function must return the wrapper

@deco
def hello():
  print("hello")
hello()
```
